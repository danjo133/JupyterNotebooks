```python
urlstr = "https://polisen.se/api/events"

data = []
with urllib.request.urlopen(urlstr) as url:
    data = json.loads(url.read().decode())

def flatten(datapoint):
    out = {}
    for key in ["datetime", "id", "name", "summary", "type", "url"]:
        out[key] = datapoint[key]
    out['county'] = datapoint["location"]["name"]
    out['gps'] = datapoint["location"]["gps"]
    return out

flattened = [flatten(x) for x in data]
df = pd.DataFrame.from_dict(flattened)
```

```python
def make_clickable(val):
    return '<a target="_blank" href="{}">{}</a>'.format(val,val)
```

```python
df2 = df.loc[df.county == "Östergötlands län"].loc[df.summary != "Information om polisens pressnummer."]
df2.loc[:,['datetime', 'name', 'type', 'url']].set_index('datetime').style.format({"url": make_clickable})
```
